```python
def to_hex(color):
    return "#{:02x}{:02x}{:02x}".format(int(color[0]*255), int(color[1]*255), int(color[2]*255))        
        
def color_curve(aa,bb,name="1",color_mode="SUMMER",arrows=False,folder="colored"):
    import matplotlib.cm as cm
    
    var("t")

    n = len(aa)

    figsize_inches = [8.27, 11.69] 
    dpi = 1000
    font_size = font_size = (figsize_inches[0] * dpi) / 200 # Example calculation
    m = min(4,n)
    ff,gg = aa,bb
    print(ff,gg)
    X = sum(ff[i]*cos(gg[i]*t) for i in range(len(ff)))+0*t
    Y = sum(ff[i]*sin(gg[i]*t) for i in range(len(ff)))+0*t

    # Number of segments
    num_segments = 400*max(1,len(ff))

    # Create a colormap using matplotlib
    cmap = cm.get_cmap(color_mode.lower())

    # Calculate the points and colors
    points = [(X(t=2*pi*i/num_segments), Y(t=2*pi*i/num_segments)) for i in range(num_segments+1)]
    colors = [cmap(i/num_segments) for i in range(num_segments)]

           

    # Create the plot
    p = Graphics()
    for ns in range(num_segments):
        # Define the segment and convert color to hex
        hex_color = to_hex(colors[ns])
        segment = line([points[ns], points[ns+1]], color=hex_color, thickness=2,axes=False,dpi=1000,
                                 #axes_labels=[str(perms[i]),group_name],
                                aspect_ratio = 1)
        if ns%100==0 and arrows:
            arr = arrow2d(points[ns],points[ns+1],color=hex_color,thickness=2,axes=False,dpi=1000,aspect_ratio=1)
            p+=arr
        p += segment

        # Show the plot
    
    #p.show(aspect_ratio=1,dpi=100)        
    p.save(folder+"/"+str(name)+".png")
    return p


```

```python
def unitary_divisors(n):
    return [d for d in divisors(n) if gcd(d,n/d)==1]


def add(u,v):
    return u*v/gcd(u,v)**2

def mat(n):
    un = unitary_divisors(n)
    return matrix([[add(u,v) for u in un] for v in un])
    


def draw_rectangle(x,y,width,height,color="black",fill=False,thickness=2):
    plt = Graphics()
    plt += polygon([(x,y),(x+width,y),(x+width,y+height),(x,y+height),(x,y)],aspect_ratio=1,dpi=1000,color=color,axes=False,fill=fill)
    plt += line([(x,y),(x+width,y),(x+width,y+height),(x,y+height),(x,y)],aspect_ratio=1,dpi=1000,thickness=thickness,color="white",axes=False)
    return plt

def regularPermutationsInSymmetricGroup(finiteGroup):
    from sage.matrix.operation_table import OperationTable
    G = finiteGroup
    O = OperationTable(G,operator.mul,names="elements")
    #print(latex(O.table()))
    ll = [ Permutation([xx +1 for xx in x]) for x in O.table()] 
    return ll

import numpy as np
import cv2 as cv
from sage.plot.colors import Color

def compute_color(i, n, color_mode=None):
    """
    Compute the i-th color from the given colormap divided into n pieces.

    Args:
    - i (int): Index of the color to compute (0 <= i <= n-1).
    - n (int): Total number of colors.
    - color_mode (str or None): The colormap to use (e.g., 'SUMMER', 'AUTUMN'). If None, uses 'SUMMER' by default.

    Returns:
    - color (Color): The computed color in SageMath's Color format.
    """
    # Prepare an array with labels 0 to n-1
    num_labels = n
    labels = np.array([[x] for x in list(range(1,n+1))])
    label_hue = np.uint8(179 * labels / np.max(labels))
    if color_mode is None:
        label_hue = cv.applyColorMap(label_hue, getattr(cv, 'COLORMAP_SUMMER'))      
    else:    
        label_hue = cv.applyColorMap(label_hue, getattr(cv, 'COLORMAP_' + color_mode.upper())) 
    
    #print(label_hue)
    # Get the color corresponding to the i-th element
    #print(len(label_hue)==n)
    lh = label_hue[i]
    
    # Extract the RGB values from OpenCV's BGR format
    b, g, r = lh[0]
    
    #print(lh)
    
    # Convert to SageMath's Color format with normalized values
    color = Color(r=r/255.0, g=g/255.0, b=b/255.0, space="rgb")
    
    return color

# Example usage:
n = 10
i = 3
color_mode = 'autumn'
color = compute_color(i, n, color_mode)
print(f"Color for i={i}, n={n}, color_mode='{color_mode}': {color}")


def tileSquare(M,color_mode=None,fill=False,thickness=2):   
    plt = Graphics()
    m = M.nrows()
    y0 = 0    
    for i in range(m):
        x0 = 0
        height = M[0,i]
        for j in range(m):
            width = M[i,j]
            color = compute_color(M[i,j]-1,m,color_mode)
            plt += draw_rectangle(x0,y0,width,height,color=color,fill=fill,thickness=thickness)
            x0 += width
        y0+=height    
    #plt.show(dpi=50)        
    return plt
N= 60
M = mat(N)

for gens,name in groups60:
    group = PermutationGroup(gens)
    M = matrix(regularPermutationsInSymmetricGroup(group))
    print(name)
    #print(M)
    for cm in ["OCEAN","AUTUMN","WINTER","SPRING","SUMMER"]:
        plt = tileSquare(M,color_mode=cm,fill=True,thickness=32.0/group.order()**1.25)
        plt.save("./latin_squares/"+str(name)+"-"+cm+".png",figsize=[8,8])
        plt.show(dpi=50)

#draw_rectangle(0,0,1,2).show(dpi=50)
```

```python

import os

def ensure_folder_exists(folder_path):
    """
    Checks if the specified folder exists. If not, creates it.

    Args:
    - folder_path (str): The path to the folder to check or create.
    
    Returns:
    - None
    """
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"Folder '{folder_path}' created.")
    else:
        print(f"Folder '{folder_path}' already exists.")     
        
import numpy as np

def draw_sigmoid_curve(p1, p2, steps=100, steepness=10):
    """
    Draw a smooth sigmoid (logistic) curve between two points.
    Args:
    - p1: Start point (x1, y1)
    - p2: End point (x2, y2)
    - steps: Number of points to sample along the curve
    - steepness: Controls the steepness of the sigmoid curve
    Returns:
    - A list of points representing the sigmoid curve.
    """
    x1, y1 = p1
    x2, y2 = p2
    
    # Sample x points between x1 and x2
    xs = np.linspace(x1, x2, steps)
    
    # Logistic sigmoid function
    def sigmoid(x):
        return 1 / (1 + np.exp(-steepness * (x - 0.5)))
    
    # Normalize x values to the range [0, 1]
    normalized_xs = (xs - x1) / (x2 - x1)
    
    # Apply the sigmoid function
    ys = sigmoid(normalized_xs)
    
    # Scale ys to fit between y1 and y2
    ys = y1 + ys * (y2 - y1)
    
    return [(x, y) for x, y in zip(xs, ys)]

def draw_permutation(x, y, size, perm, color_mode=None,steepness=10,thickness=2):
    plt = Graphics()
    n = len(perm)
    num_labels = n**2
    labels = np.array([[x] for x in list(range(0,num_labels))])
    label_hue = np.uint8(179 * labels / np.max(labels))
    if not color_mode is None:
        label_hue = cv.applyColorMap(label_hue, getattr(cv, 'COLORMAP_' + color_mode.upper()))
    else:
        label_hue = cv.applyColorMap(label_hue, getattr(cv, 'COLORMAP_SUMMER'))    
    
    
    #print(perm)
    for i in range(1, n + 1):
        j = perm[i - 1]
        start_point = (x, y + i / (n + 1) * size)
        end_point = (x + size, y + j / (n + 1) * size)
        sigmoid_curve = draw_sigmoid_curve(start_point, end_point, steepness=steepness)
        if color_mode is None:
            color = "black"
        else:
            #h = 
            cl = (i)*(j)
            lh = label_hue[int(cl/(num_labels)*(len(label_hue)-1))]
            #print(cl,lh)
            b,g,r = lh[0]
            color = Color(r=r/255,g=g/255,b=b/255,space="rgb")            
        plt += line(sigmoid_curve, dpi=1000, aspect_ratio=1, axes=False, color=color,thickness=thickness)
    
    return plt

def draw_group(group, name, x=0, y=0, size=1, color_mode=None, steepness=10,thickness=4):
    n = group.order()
    perms = regularPermutationsInSymmetricGroup(group)
    plt = Graphics()
    
    
    for i in range(n):
        for j in range(n):
            x0, y0 = x + size * j, y + size * (n-1-i)
            #print(x0,y0,perms[i]*perms[j])        
            plt += draw_permutation(x0, y0, size=size, perm=perms[i] * perms[j], color_mode=color_mode, steepness=steepness,thickness=thickness)
    folder = "./finite_group_brezels/"
    ensure_folder_exists(folder)
    cm = "bw" if color_mode is None else color_mode
    plt.save(folder+name+"_"+cm+".png",figsize=[8,8],dpi=1000)            
    return plt        

# Example usage with SymmetricGroup(3) and logistic (sigmoid) curves
for gens,name in groups60[0:50]:
    group = PermutationGroup(gens)
    print(name)
    for cm in [None,"AUTUMN","SPRING","SUMMER","WINTER"]: 
        print(cm)
        plt = draw_group(group,name, color_mode = cm,steepness=10,thickness=32/group.order()**(1.25))
        plt.show(dpi=50,aspect_ratio = 1)
```
